```python
import wordcloud
from matplotlib import pyplot as plt
```

The next section saves all the txt in a list, every line of the txt is an element of the list. \
To open your file copy the direction of the .txt file and paste it in 'Paste the direction file here'

```python
with open(r'Paste the direction file here', 'r', encoding='utf-8') as file:
    file_contents = file.readlines() #Read all the txt lines
```

The next function counts all the words of the txt file that are not in the array 'uninteresting_words' without taking into account the punctuations

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just", "for", "in", "not"]
         
    # LEARNER CODE START HERE
    frequencies = {} #Words counting
    for line in file_contents: #As file_contents has every line as an object of the list, we iterate thorugh each item
        for word in line.lower().split(): # Every item is a hole sentence so we split it into words in lower case
            cleaned_word = '' #Some words may have a punctuation attached so we must clean it in a new word
            for letter in word: #Iterate through each letter of the current word
                if letter not in punctuations: #Check if the letter is not a punctuation
                    cleaned_word += letter #If not a new word is created so it wont consider any punctuation
            if cleaned_word not in uninteresting_words: #If the word is not in the above list
                if cleaned_word not in frequencies.keys(): #Check the entry in the dictionary 
                    frequencies[cleaned_word] = 0 # Create the entry
                frequencies[cleaned_word] += 1 #Sum the word
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(frequencies) #Generate the word cloud with the frequencies dictionary
    return cloud.to_array(), frequencies
```

```python
# Display your wordcloud image
myimage, dictionary = calculate_frequencies(file_contents) #Call the function
plt.imshow(myimage, interpolation = 'nearest') #Show the image
plt.axis('off')
plt.show()
```
